# Section 1: Theory

**1. In your own words, what does the role of a data scientist involve?**

The role of a data scientist involves leveraging data to *extract valuable insights* and *make informed decisions*. 

Data scientists are responsible for collecting, cleaning, and analysing large datasets using statistical techniques and machine learning algorithms. They develop models and algorithms to solve complex problems, identify patterns and trends, and communicate their findings to stakeholders through data visualisations and reports. Additionally, data scientists often collaborate with cross-functional teams to address business challenges, optimise processes, and drive data-driven decision-making.

**2. What is an outlier? Here we expect to see the following:
  a. a definition
  b. Examples
  c. Should outliers always be removed? Why?
  d. What are other possible issues you can find in a dataset?**

a. An outlier is a data point that significantly deviates from the overall pattern or distribution of the dataset. It is an observation that lies an abnormal distance away from other observations.

b. Examples of outliers:

In a dataset of students' ages, if there is an entry showing an age of 150 years, it would be considered an outlier.
In a dataset of housing prices in a neighborhood, if there is a house listed with an exceptionally high or low price compared to other houses, it would be an outlier.

c. Outliers do not always need to be removed. Whether to remove outliers or not depends on the context and the goal of the analysis. Sometimes outliers represent valid and important information, such as rare events or extreme conditions. In such cases, removing outliers could lead to a loss of valuable insights. However, in certain situations where outliers are caused by errors, measurement issues, or data anomalies, removing them may improve the accuracy and reliability of the analysis.

d. Other possible issues that can be found in a dataset include:

- Missing data: Some entries or values may be missing, which can introduce bias or affect the validity of the analysis.

- Data inconsistencies: Inconsistent or contradictory values within the dataset can lead to inaccurate conclusions.

- Skewed distributions: Skewed or non-normal distributions can affect the assumptions of statistical methods and may require appropriate transformations or alternative approaches.

**3. Describe the concepts of data cleaning and data quality.Here we expect to see the following: a. What is data cleaning? b. Why is data cleaning important? c. What type of mistakes do we expect to commonly see in datasets?**

a. Data cleaning refers to the process of identifying and correcting or removing errors, inconsistencies, and inaccuracies in a dataset. It involves various activities such as handling missing data, correcting erroneous values, standardizing formats, resolving inconsistencies, and removing duplicate entries.

b. Data cleaning is important for several reasons:

- Accurate analysis: Clean data ensures that the analysis and insights drawn from the data are reliable and trustworthy.
- Reliable decision-making: Clean data supports informed decision-making by providing accurate and consistent information.
- Improved data quality: Cleaning data helps improve the overall quality of the dataset, making it more valuable and usable for future analyses.
- Effective data modeling: Clean data is essential for building accurate and robust predictive models.

c. Common mistakes or issues commonly found in datasets include:

- Missing values: Some entries may have missing data, which can lead to biased or incomplete analyses.
- Data entry errors: Typos, transcription errors, or incorrect values during data entry can introduce inaccuracies in the dataset.
- Inconsistent formats: Inconsistent formatting of dates, addresses, or categorical variables can hinder analysis and comparisons.
- Duplicate entries: Duplicated rows or records can skew analysis results and lead to incorrect conclusions.
- Outliers: Outliers are data points that deviate significantly from the rest of the dataset and may distort analysis results if not properly handled.
- Inaccurate or outdated data: Data that is not up-to-date or contains outdated information can impact the validity and relevance of the analysis.

**4. Discuss what is Unsupervised Learning - Clustering in Machine Learning using an example. Here we expect to see the following:
a.Definition.
b.When is it used?
c.What is a possible real-world application of unsupervised learning?
d.What are its main limitations?**

a. Unsupervised Learning - Clustering in Machine Learning:
Unsupervised learning is a branch of machine learning where the goal is to discover patterns, structures, or relationships in a dataset without any predefined labels or target variables. Clustering is a popular technique in unsupervised learning that *groups similar data points together based on their inherent similarities or proximity*.

b. When is it used?
Unsupervised learning and clustering are used in various scenarios, including:

- Exploratory data analysis: Clustering helps in understanding the underlying structure and patterns within a dataset when there is no prior knowledge about the data.
- Customer segmentation: Clustering techniques can be used to segment customers into distinct groups based on their purchasing behavior, demographics, or preferences.
- Image or document categorization: Clustering can be employed to automatically group similar images or documents together based on their content or features.
- Anomaly detection: Clustering can identify unusual or anomalous patterns in data, helping in detecting fraud, network intrusions, or other abnormal behaviors.

c. Real-world application of unsupervised learning:
An example of a real-world application of unsupervised learning and clustering is market segmentation in the retail industry. By analysing customer purchasing behaviour, demographic data, and other relevant factors, clustering algorithms can group customers into segments based on their similarities. This information can then be used to personalise marketing campaigns, optimise product offerings, and tailor customer experiences.

d. Limitations of unsupervised learning and clustering:

- Clustering results can be subjective as there is *no ground truth or predefined labels*. Different clustering algorithms or parameter settings may yield different results, and the interpretation of clusters is often based on human judgment.
- Deciding on the appropriate number of clusters in a dataset is challenging and often requires trial and error or domain expertise.
- Clustering results can be sensitive to the scale, normalisation, or preprocessing of the input data. Outliers or irrelevant features can also impact the clustering outcome.
- Difficulty in handling high-dimensional data: Clustering algorithms may struggle to effectively handle datasets with a high number of features or dimensions, often referred to as the "curse of dimensionality."
- Lack of interpretability: Clustering results may lack interpretability, especially in complex datasets, making it challenging to derive actionable insights without additional analysis or domain knowledge.

**5. Discuss what is Supervised Learning - Classification in Machine Learning using an example. Here we expect to see the following:
a.Definition.
b.When is it used?
c.What is a possible real-world application of supervised learning?
d.What data do we need for it? Is there any processing that needs to be done?**

a. Supervised learning is a branch of machine learning where the goal is to learn a mapping function from input features to output labels based on labeled training data. Classification, a common task in supervised learning, involves predicting the class or category of a given input based on its features and the training examples with known labels.

b. Supervised learning and classification are used in various scenarios, including:

- Spam email detection: Classifying emails as either spam or non-spam based on their content and features.
- Sentiment analysis: Classifying text documents or social media posts as positive, negative, or neutral based on their sentiment.
- Image recognition: Identifying objects or patterns in images and classifying them into predefined categories.
- Disease diagnosis: Predicting the presence or absence of a disease based on patient symptoms, medical test results, or genetic markers.

c. An example of a real-world application of supervised learning and classification is credit card fraud detection. By training a classification model on historical credit card transactions, where the fraud label is known, the model can learn patterns and features indicative of fraudulent transactions. This trained model can then be used to predict whether new transactions are likely to be fraudulent or not.

d. Supervised learning requires *labeled training data*, which consists of input features and their corresponding output labels. The input features represent the characteristics or attributes of the data, while the output labels represent the target variable or the class/category to be predicted.

Before using the data for supervised learning, some preprocessing steps may be necessary, including:

- Handling missing values: Missing data may need to be imputed or removed to ensure the integrity of the dataset.
- Feature scaling: Scaling or normalising the input features can help bring them to a similar scale, preventing certain features from dominating the learning process.
- Splitting the data: The labeled data is typically split into training and testing sets. The training set is used to train the model, while the testing set is used to evaluate its performance on unseen data.

The preprocessing steps may vary depending on the specific dataset and the requirements of the classification problem at hand.
